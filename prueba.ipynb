```python
from random import choice, randint
from datetime import date

def generar_datos_sinteticos():
    # Lista de posibles valores para tx_rt
    tx_rt = ['000112', '000113', '000114']
    
    # Nombre
    nombre = ["Marcos","Jose","Maria"]
    
    # País
    pais =  ["Peru","EEUU","España","Malta","Australia","Portugal","Rumania"]
    
    # Fecha de nacimiento
    # Se utiliza un rango de fechas desde 2025 hasta 2100
    anio_min = 1950
    anio_max = 2024
    dia = randint(1, 28)  # Considera que los meses con menos de 31 días
    mes = randint(1, 12)
    fec_nacimiento = f"{dia:02d}/{mes:02d}/{randint(anio_min, anio_max)}"
    
    return {
        "tx_rt": choice(tx_rt),
        "nombre": choice(nombre),
        "pais": choice(pais),
        "fec_nacimiento": fec_nacimiento
    }

# Ejemplo de uso
print(generar_datos_sinteticos())
```

```python
from typing import List

def analisis_numeros(numeros: List[float]) -> dict:
    """
    Analiza una lista de números y devuelve un diccionario con el promedio, valor máximo,
    valor mínimo y cantidad de elementos. Maneja listas vacías y valores no numéricos.
    
    :param numeros: Lista de números (puede incluir None, que se tratará como 0).
    :type numeros: List[float]
    :return: Diccionario con 'promedio', 'maximo', 'minimo' y 'cantidad'.
    :rtype: dict
    """
    
    if not numeros:
        return {
            'promedio': None,
            'maximo': None,
            'minimo': None,
            'cantidad': 0
        }
    
    maximo = -float('inf')
    minimo = float('inf')
    cantidad = 0
    suma = 0
    
    for numero in numeros:
        if numero is None:
            numero = 0
        if not isinstance(numero, (int, float)):
            continue
        cantidad += 1
        suma += numero
        if numero > maximo:
            maximo = numero
        if numero < minimo:
            minimo = numero
    
    promedio = suma / cantidad if cantidad > 0 else None
    
    return {
        'promedio': promedio,
        'maximo': maximo,
        'minimo': minimo,
        'cantidad': cantidad
    }
```

```python
analisis_numeros([1,2,4,5])
```
